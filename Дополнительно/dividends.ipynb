Анализ дивидендов

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def get_dividends(ticker):
    stock = yf.Ticker(ticker)
    dividends = stock.dividends
    return dividends

def plot_dividends(dividends, ticker):
    if dividends.empty:
        return None
    plt.figure(figsize=(10, 6))
    dividends.plot(label='Дивиденды', color='green')
    plt.title(f'дивиденды по акциям {ticker}')
    plt.xlabel('Дата')
    plt.ylabel('Сумма дивидендов (USD)')
    plt.legend()
    return plt
```

```python
ticker = 'AAPL'
dividends = get_dividends(ticker)
plot_dividends(dividends, ticker)
```
