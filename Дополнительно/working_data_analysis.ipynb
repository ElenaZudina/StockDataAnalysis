```python
def plot_volume_trend(data, ticker):
    plt.figure(figsize=(10, 6))
    data['Volume'].plot(label='Объем торгов', color='blue')
    plt.title(f'График объема торгов {ticker}')
    plt.xlabel('Дата')
    plt.ylabel('Объем торгов')
    plt.legend()
    return plt
```

```python
stock = yf.Ticker('AAPL')
data = stock.history(period='1mo')

plot_volume_trend(data, 'AAPL')

plt.show()
```
